```python
from dotenv import load_dotenv
import os
import boto3
from pathlib import Path

load_dotenv()
```

```python
bucket = 'carbon-dioxide-data'
endpoint_url = os.getenv("B2_ENDPOINT_URL")
region = os.getenv("B2_REGION")

aws_access_key_id = os.getenv("B2_KEY_ID")
aws_secret_access_key = os.getenv("B2_APP_KEY")

PROJECT_ROOT = Path.cwd().parent
local_data_dir = PROJECT_ROOT / "data" / "processed"
local_model_dir = PROJECT_ROOT / "models"
```

```python
s3 = boto3.client(
    "s3",
    endpoint_url=endpoint_url,
    aws_access_key_id=aws_access_key_id,
    aws_secret_access_key=aws_secret_access_key
)
```

```python
def upload_file(local_path: Path, s3_key: str):
    if not local_path.exists():
        print(f"❌ File not found: {local_path}")
        return
    print(f"⬆️ Uploading {local_path} → b2://{bucket}/{s3_key}")
    s3.upload_file(str(local_path), bucket, s3_key)
```

```python
# ---- Upload required datasets ----
upload_file(local_data_dir / "test_feature_engineered.csv", "processed/test_feature_engineered.csv")
upload_file(local_data_dir / "test_cleaned.csv", "processed/test_cleaned.csv")
upload_file(local_data_dir / "train_feature_engineered.csv", "processed/train_feature_engineered.csv")

# ---- Upload model ----
upload_file(local_model_dir / "xgb_best_model.pkl", "models/xgb_best_model.pkl")
```
