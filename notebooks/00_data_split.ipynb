```python
df = pd.read_csv(os.path.join('..', 'data', 'raw', 'carbon_dioxide_emissions.csv'), skiprows=4)
```

```python
df.drop(columns=['Indicator Name', '1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969', '2025', 'Unnamed: 70'], inplace=True)
```

```python
# Train set: 

# identify year columns
year_cols = [c for c in df.columns if c.isdigit()]

# long format
long_df = df.melt(
    id_vars=['Country Name', 'Country Code', 'Indicator Code'],
    value_vars=year_cols,
    var_name='Year',
    value_name='CO2_Mt'
)

long_df['Year'] = long_df['Year'].astype(int)

# pivot to years as rows, countries as columns
wide_year_country = long_df.pivot_table(
    index='Year',
    columns=['Country Name', 'Country Code', 'Indicator Code'],
    values='CO2_Mt'
)

wide_year_country.head()
```
