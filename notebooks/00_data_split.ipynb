```python
import matplotlib.pyplot as plt
import os
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

### Prepare Carbon Dioxide Emissions Data

```python
df_carbon = pd.read_csv(os.path.join('..', 'data', 'raw', 'carbon_dioxide_emissions.csv'), skiprows=4)
```

```python
df_carbon.shape
```

```python
df_carbon.head()
```

```python
df_carbon.drop(columns=['Country Code', 'Indicator Code', '2025', 'Unnamed: 70'], inplace=True)

df_carbon = df_carbon.melt(
    id_vars=["Country Name", "Indicator Name"],
    var_name="Year",
    value_name="Value"
)
```

```python
df_carbon.sample(10)
```

---
### Prepare World Development Indicators Data

```python
df_indicators = pd.read_csv(os.path.join('..', 'data', 'raw', 'world_development_indicators.csv'), skiprows=4)
```

```python
df_indicators.shape
```

```python
df_indicators.head()
```

```python
df_indicators.drop(columns=['Country Code', 'Indicator Code', '2025', 'Unnamed: 70'], inplace=True)

df_indicators = df_indicators.melt(
    id_vars=["Country Name", "Indicator Name"],
    var_name="Year",
    value_name="Value"
)
```

```python
df_indicators.sample(20)
```

---
### Merge Data

```python
df_merged = pd.concat([df_carbon, df_indicators])
```

```python
df_merged.sample(20)
```

```python
df_merged["Year"] = df_merged["Year"].astype(int)
df_merged["Value"] = pd.to_numeric(df_merged["Value"], errors="coerce")
```

```python
df_merged.sample(20)
```

```python
df_final = df_merged.pivot_table(
    index=["Country Name", "Year"],
    columns="Indicator Name",
    values="Value"
).reset_index()
```

```python
df_final.head(10)
```

---
### Split Data

```python
cutoff_year_test = 2009
cutoff_year_holdout = 2021
```

```python
train_df = df_final[df_final['Year'] < cutoff_year_test]
test_df = df_final[(cutoff_year_test <= df_final['Year']) & (df_final['Year'] < cutoff_year_holdout)]
holdout_df = df_final[cutoff_year_holdout <= df_final['Year']]

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)
print("Holdout shape:", holdout_df.shape)
```

```python
train_df.to_csv(os.path.join('..', 'data', 'raw', 'train.csv'), index=False)
test_df.to_csv(os.path.join('..', 'data', 'raw', 'test.csv'), index=False)
holdout_df.to_csv(os.path.join('..', 'data', 'raw', 'holdout.csv'), index=False)
```
